```python
# функция генерации выборки
def Pois(lamb, n):
    nums = []
    for i in range(n):
        p = math.exp(-lamb)
        sp = 0
        sp += p
        u = random.uniform(0,1)
        j = 0
        while u > sp:
            j += 1         
            p *= lamb / j  
            sp += p
        nums.append(j)
    return nums
```

```python
sample5=[]
sample100=[]
sample1000=[]
sample5000=[]
sample105 = []
for i in range(5):
    sample5.append(Pois(5, 5))
    sample1000.append(Pois(5, 1000))
    sample5000.append(Pois(5, 5000))
    sample105.append(Pois(5, 10**5))
    sample100.append(Pois(5, 100))
```

```python
quant = {0.9: 0.81594, 0.95: 0.828944}

def l_part(distr, par):
    npar = par * len(distr)
    return (sum(distr) - npar) / np.sqrt(npar)

def print_nptest(distr, params, alpha):
    lp = l_part(distr, params)
 
    ta = quant[1-alpha]
    h0 = True if lp < ta else False
    print(f'Левая часть равна {np.round(lp, 3):6}, t_alpha = {ta}'\
    f'. Л.ч {"<" if h0 else ">"} t_alpha => гипотеза H0 {"не" if not h0 else "  "} принимается')
```

```python
distrs_v = [sample5, sample100, sample1000, sample5000, sample105]
params = 5
alphas = [0.1, 0.05]
for distrs in distrs_v:
    print(f'\n================================== Для выборок объемом {len(distrs[0]):6} =================================')
    for alpha in alphas:
        print(f'--------------------------------------- Для alpha = {alpha:4} --------------------------------------')
        for distr in distrs:
            print_nptest(distr, params, alpha)
```

```python
sample5, sample100, sample1000, sample5000, sample105 = [], [], [], [], []
for i in range(5):
    sample5.append(Pois(1, 5))
    sample1000.append(Pois(1, 1000))
    sample5000.append(Pois(1, 5000))
    sample105.append(Pois(1, 10**5))
    sample100.append(Pois(1, 100))
    
distrs_v = [sample5, sample100, sample1000, sample5000, sample105]
params = 1
alphas = [0.1, 0.05]
for distrs in distrs_v:
    print(f'\n================================== Для выборок объемом {len(distrs[0]):6} =================================')
    for alpha in alphas:
        print(f'--------------------------------------- Для alpha = {alpha:4} --------------------------------------')
        for distr in distrs:
            print_nptest(distr, params, alpha)
```
