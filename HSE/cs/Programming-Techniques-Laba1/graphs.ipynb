```python
import pandas
# подключение модуля и сокращение названия для удобства
import pandas as pd

# для использования функций numpy
import numpy as np

#для отрисовки графиков
import matplotlib.pyplot as plt
%matplotlib inline

filename="D:\\my_folder\\jupyter\\laba1\\sorting\\cmake-build-debug\\time.csv"
df = pd.read_csv(filename,skiprows=0)
df.head()
ans=pd.pivot_table(df, index='index')
ans
```

```python
temp_plot = ans.plot(figsize=(10, 8), logy=True)
temp_plot.set_ylabel('Time, microseconds', fontsize=15)
temp_plot.set_xlabel('size', fontsize=15)
temp_plot.set_title('Logarithmic plot of sorts timing', fontsize=15)
plt.savefig('timing_graph.png')
# temp_plot.plot(loglog=True)
```
